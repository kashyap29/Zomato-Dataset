```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# Pie chart - top 3 counrty that uses zomatos
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation : Zomato maximum records or transaction are from India, After that USA and UK


```python
# Relationships
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

# Observation 
1. when rating is between 4.5 to 4.9 ----> Excellent
2. when rating is between 4..0 to 4.4 ----> very Good
3. when rating is between 3.5 to 3.9 ----> Good
4. when rating is between 2.5 to 3.4 ----> Average
5. when rating is between 1.8 to 2.4 ----> Poor
6. when rating is between 0.0  ----> Not rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette = ['Orange','red','orange','yellow','green','green'])
```

observation:
1. Not rated count is very high
2. Max number of ratings are between 2.5 and 3.4

```python
#Count plot
sns.countplot(x='Rating color', data= ratings, palette = ['Orange','red','orange','yellow','green','green'])
```
